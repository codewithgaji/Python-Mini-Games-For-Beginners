# Coin Toss Simulator


```python
import random

def coin_toss():
    print("Welcome to Gaji's Coin Toss Game :)")
    print("You and Stark get to go on 5 exciting rounds to see who wins!")

    # Variables
    coin_face = ['Heads', 'Tails']
    player_score = 0
    stark_score = 0
    flip_count = 0
    heads = 0
    tails = 0
    max_attempts = 5

    for attempt in range(1, max_attempts + 1):
        try:
            # Player chooses Heads or Tails
            q1 = input("\nWhat do you choose? (Heads/Tails): ").lower()
            if q1 not in ["heads", "tails"]:
                print("Invalid input. Please choose 'Heads' or 'Tails'.")
                continue
            
            # Stark automatically chooses the opposite
            if q1 == 'heads':
                print("Stark: I choose tails.")
            else:
                print("Stark: I choose heads.")

            # Player decides to flip or not
            flip_choice = input("\nWould you like to flip the coin? (yes/no): ").lower()
            if flip_choice not in ["yes", "no", "yup", "yeah", "sure", "okay"]:
                print("Invalid input. Please type 'yes' or 'no'.")
                continue

            if flip_choice in ["no"]:
                print("\nGame ended early!")
                break

            # Toss the coin
            print("Flipping...")
            toss = random.choice(coin_face)
            print(f"The coin landed on {toss}.")

            # Update scores based on the toss result
            flip_count += 1
            if toss.lower() == "heads":
                heads += 1
            else:
                tails += 1

            if toss.lower() == q1:
                print("You win this round!")
                player_score += 1
            else:
                print("Stark wins this round!")
                stark_score += 1

            # Display current scores
            print(f"\nCurrent Score: Stark: {stark_score}, You: {player_score}")

        except ValueError:
            print("Error: Please follow the instructions!")

    # Final game results
    print("\nGame Over!")
    print(f"You flipped the coin {flip_count} times.")
    print(f"You tossed {heads} heads and {tails} tails.")

    if player_score > stark_score:
        print("Congratulations, You Win Overall! :)")
    elif player_score < stark_score:
        print("Stark Wins Overall! :(")
    else:
        print("It's a Draw!")

# Start the game
coin_toss()
```
